```python
from bert_sequence_vectorizer import BertSequenceVectorizer
import pandas as pd

df = pd.read_pickle('onsen_dataset.pkl')
LEAST_REVIEWS = 5

groups = df.groupby('onsen')
print('all onsen:', len(groups))

onsen_more_than_reviews = []
for onsen, group in groups:
    if len(group) > LEAST_REVIEWS:
        onsen_more_than_reviews.append(onsen)
print(str(LEAST_REVIEWS) + '> onsen:', len(onsen_more_than_reviews))
df = df[df['onsen'].isin(onsen_more_than_reviews)]
df = df[df['is_review'] == False]
df
```

```python
BSV = BertSequenceVectorizer()
df['text_feature'] = df['comment'].progress_apply(lambda x: BSV.vectorize(x))
df
```

```python
import numpy as np
from bert_sequence_vectorizer import cos_sim_matrix

mat = cos_sim_matrix(np.stack(df.text_feature))
mat
```

```python
df_r = df.reset_index()
idx = df_r[df_r.onsen == '城崎温泉'].index.tolist()[0]
cos_d = mat[idx]
result_idx = pd.Index(np.argsort(cos_d)[::-1][:10])
df_r.iloc[result_idx]
```

「浴衣」と「ゆかた」が同じ意味の言葉として認識できているので「かみのやま温泉」を提案できている．

Doc2Vecによる学習時は「浴衣」と「ゆかた」が同じ意味として捉えられるほど学習データがなかったため，「城崎温泉」と「かみのやま温泉」は近いベクトルにならなかったのだろうと考えられる．
